```python
import itertools
import re
from collections import OrderedDict
source = "mushroom.txt"
```

```python
def naturalJoin(level):
    ret = []
    size = len(level[0])
    for i in range(len(level)):
        for j in range(i+1, len(level)):
            if level[i][:-1] == level[j][:-1]:
                new = sorted(set(level[i]) | set(level[j]))
                ret.append(new)
    return ret


def prune(ck, lk_1):
    ret = []
    for itemSet in ck:
        valid = True
        subSets = list(itertools.combinations(itemSet,len(itemSet)-1))
        for subSet in subSets:
            if(list(subSet) not in lk_1):
                valid = False
                break
        if valid:
            ret.append(itemSet)
    return ret


def newLevel(ck,transactions):
    ret = []
    for i in range(len(ck)):
        cnt = 0
        for j in range(len(transactions)):
            if(set(ck[i])<=set(transactions[j])):
                cnt+=1
        if(cnt>=min_sup):
            ret.append(ck[i])
    return ret
```

```python
import math
dataSource = open(source,"r")
transactions = []
for line in dataSource:
    tmp = [int(i) for i in line.split()]
    transactions.append(tmp)
candidates = {}
sup = int(input("Enter minimum support percentage: "))
min_sup = math.ceil((sup/100)*len(transactions))
print(min_sup)

import time
start_time = time.time()
cnt = 0
for i in range(len(transactions)):
    for j in range(len(transactions[i])):
        if transactions[i][j] not in candidates:
            candidates[transactions[i][j]] = 1
            cnt += 1
        else:
            candidates[transactions[i][j]] += 1
            

dataDictionary = OrderedDict(sorted(candidates.items()))
cdList = []
cdList.append(list(dataDictionary.keys()))

l1 = []
lvlList = []
for dataKey, fre in dataDictionary.items():
    if(fre >= min_sup):
        l1.append([dataKey])
# print("L1: ",end='')
lvlList.append(l1)
sumjoin = 0
sumprune = 0
sumFP = 0
nw = start_time
# print(" After joining: "+str(cnt),end='')
# print(" After pruning: "+str(cnt),end='')
# print(" After FP: "+str(len(l1)),end='')
# print(" Required time: "+str(time.time()-nw))
nw = time.time()
for k in range(1,300):
#     print("L"+str(k+1)+": ",end='')
    tmp = naturalJoin(lvlList[k-1])
    if(len(tmp) == 0):
        break
#     print(" After joining: "+str(len(tmp)),end='')
    sumjoin += len(tmp)
    tmp = prune(tmp, lvlList[k-1])
    if(len(tmp) == 0):
        break
#     print(" After pruning: "+str(len(tmp)),end='')
    sumprune += len(tmp)
    cdList.append(tmp)
    tmp = newLevel(tmp, transactions)
#     print(" After FP: "+str(len(tmp)),end='')
    sumFP += len(tmp)
#     print(" Required time: "+str(time.time()-nw))
    if(len(tmp) == 0):
        break
    else:
        lvlList.append(tmp)
    nw = time.time()

# print("SumofJoin: "+str(sumjoin))
# print("SumofPrune: "+str(sumprune))
# print("SumofFP: "+str(sumFP))
print("Total Time Taken: "+ str(time.time()-start_time))
```
